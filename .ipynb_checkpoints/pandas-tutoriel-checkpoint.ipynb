## Changer nom des colonnes

`DataFrame.rename(self, mapper=None, index=None, columns=None, axis=None, copy=True, inplace=False, level=None, errors='ignore')`

```python
# Ancien_ID pour Ancien ID
df = pd.DataFrame({"ID": [100041114, 100002880, 100003032], "Ancien_ID": ["NaN", 10.0, 3.0], "Nom": ["Eco-Display Parc Stanley", "Pont Jacques-Cartier", "Berri"]})
df.rename(columns={"Ancien_ID": "Ancien ID"})
```

```python
df[:3].rename(index={0: "Premier", 1: "Deuxieme", 2: "Troisieme"})
```

```python
df[:3].rename(str.lower, axis='columns')
```

```python
df[:3].rename({0: 1, 1: 2, 2: 3}, axis='index')
```

## Remplacer des chaînes de caractères
`Series.str.replace(pat, repl, n=-1, case=None, regex=True)`

```python
df.iloc[27]
```

```python
df["Nom"] = df["Nom"].str.replace('è', 'e')
df["Nom"] = df["Nom"].str.replace(',', 'e')
df["Nom"] = df["Nom"].str.replace('_', ' ')

df
```
