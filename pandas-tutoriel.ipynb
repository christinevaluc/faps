### Exemple
Écrire un algorithme qui retourne le nom des compteurs de vélos actifs contenu dans le fichier `localisation_des_compteurs_velo.csv`.

```python
import pandas as pd

actif = []

data = pd.read_csv('./donnees-ouvertes/localisation_des_compteurs_velo.csv', delimiter=',', index_col="Nom")

statut = data["Statut"]

dict_stat = statut.to_dict()

for cle, valeur in dict_stat.items():
    if valeur == 'Actif':
        actif.append(cle)

print(actif)
```
