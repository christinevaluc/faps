# R vs Python

## Opérateurs
Les opérations arithmétiques comme l'addition, la substraction, la multiplication, la division et l'exponentiation est pareil dans les deux languages.

```python
sum = 1 + 3*4 + 5/2 + 2**2
print(sum)
```

Les opérateurs de comparaison sont aussi identiques.

```python
res = 5 > 3
print(res)
print(res == False)
print(res != True)
```

## Data Types
**Rappel** : Pour créer une nouvelle variable sur R, on utilise `<-` pour désigner l'affection d'une variable.

Sur Python, on utilise `=` pour affecter une variable. De plus, une variable n'est qu'un seul objet unique et ne peut contenir des sous-variables.

```python
mydata1 = 5
mydata2 = 2

mydata_sum = mydata1 + mydata2
print(mydata_sum)
```

Sur Python, une fonction est un bloc de code qui permet de découper un programme en plusieurs sous-programmes. Elle contient un nom, des paramètres (optionnels) et une valeur de retour (optionnelle).

```python
def nom_de_ma_fonction(paramètres):
    # bloc d'instructions
    res = "Bonjour " + paramètres
    return res
```
